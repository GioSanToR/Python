# **Kaprekar's Routine**

```python
input_number = 4157 #@param {type:"integer"}
```

```python
number = str(input_number)
beta_list = sorted(number)
alpha_list = sorted(number,reverse=True)
alpha = ''.join(alpha_list)
beta = ''.join(beta_list)
K = str(int(alpha) - int(beta))
Kaprekar_seq =[number]
while number != K and Kaprekar_seq[0] != K:
    if K == 0:
        Kaprekar_seq += [0]
        break
    else:
        number = K
        beta_list = sorted(K)
        alpha_list = sorted(K,reverse=True)
        alpha = ''.join(alpha_list)
        beta = ''.join(beta_list)
        Kaprekar_seq += [K]
        K = str(int(alpha) - int(beta))
print(Kaprekar_seq)
```
