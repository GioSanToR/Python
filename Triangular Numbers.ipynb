## **Trinagular numbres which are the product of three consecutive numbers**

```python
input_number = 100000000 #@param {type:"integer"}
```

```python
number = input_number
for i in range (1,number):
    x = 8*i*(i+1)*(i+2) +1
    y = x**0.5
    if y.is_integer():
        n = int((y - 1)/2)
        Tn = int(n*(n+1)/2)
        if Tn ==  i*(i+1)*(i+2):
            print(f"the {n}th triangular number {Tn} is the product of {i},{i+1},{i+2}")
```
