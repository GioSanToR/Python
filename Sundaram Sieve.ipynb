# **Sundaram Sieve to calculate primes less than a number**

```python
input_number = 1000 #@param {type:"integer"}
```

```python
def sundaram(N):
    numberList = [] 
    primeList = []
    for i in range(1,int(N/3)):
        for j in range (1,int(N/3)):
            number = i + j + 2*i*j
            if number <= N:
                numberList.append(number)       
    for i in range(1,N):
        if i not in numberList:
            prime = 2*i +1
            primeList.append(prime)
    return primeList
```

```python
print(sundaram(int(input_number/2)))
```
