# **Super-Heronian triangles**

Super-Heronian triangle to be a triangle with sides three consecutive integers and whose area is also an integer. An example of a Super-Heronian triangle is the 3, 4, 5 right triangle with area 6

```python
from decimal import Decimal

for n in range(2,10000000,2):
    number1= n**2 - 4
    number2=(number1/3)**0.5
    if Decimal(number2).as_integer_ratio()[1] == 1:
        print(f'The {[n-1,n, n+1]} Super-Heronian Triangle has area {int(n/4*(3*number1)**0.5)}')
        
```
