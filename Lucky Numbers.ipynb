# Lucky Numbers

```python
#@title Lucky Numbers less than the input number
number = 156 #@param {type:"integer"}
```

```python
dictnumbers = {}
N = number
for i in range (1,N,2):
    dictnumbers[i] = True #creates a dictionary with odd positive integers as keys and True as values
n = 2
iterations = int(0.03*N + 6) #number of minimal iterations is almost linear
while n < iterations: # (N, minimal iterations) = (50,6), (100,8), (500,21),(1000,33), (5000,100), (10000,170), (20000,284), (30000,387)
    for j in range (list(dictnumbers)[n-1]-1,len(dictnumbers),list(dictnumbers)[n-1]): 
        dictnumbers[list(dictnumbers)[j]] = False #changes to false the values according to the Lucky numbers sieve
    for x in range (max(dictnumbers.keys())+1):
        if dictnumbers.get(x) == False:
            dictnumbers.pop(x) #deletes keys with False values
    n += 1
luckynumbers = list(dictnumbers.keys())
print(luckynumbers)
print(f'There are {len(luckynumbers)} lucky numbers less than {N}')
```
