# **Lychrel Numbers and Palindromes**

```python
#@title Write a number
input_number = 689878 #@param {type:"integer"}
```

```python
number = str(input_number)

def invert(n):
    invert_n = str()
    for i in range(len(n)):
        invert_n +=  (n[len(n)-i-1])
    return invert_n

if number != invert(number):
    number_p = number
    not_palindrome = True
    k = 1
    while not_palindrome:
        new_number = str(int(number_p) + int(invert(number_p)))
        if new_number == invert(new_number):
            print(f"{number} reaches a palindrome {new_number} in {k} iterations")
            not_palindrome = False
        else:
            number_p = new_number
            k += 1
            if k == 300:
                print(f"{number} is possibly a Lychrel number")
                not_palindrome = False
else:
    print(f"{number} is a palindrome")
```
