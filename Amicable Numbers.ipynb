# **Amicable Numbers and Aliquot Sequences**

```python
def aliquotSum(n):
    import math
    s = 1
    k = 2
    if (n**0.5).is_integer():
        s = 1 + int(n**0.5)
        while k < int(n**0.5):
            if n % k == 0:
                s += k + int(n/k)
            k += 1        
    else:
        while k <= math.floor(n**0.5):
            if n % k == 0:
                s += k + int(n/k)
            k += 1
    return s
```

```python
def amicable(limit):
    pairs = []
    for i in range(2,limit):
        p = aliquotSum(i)
        if p != i:
            q = aliquotSum(p)
            if q == i and p > q:
                pairs.append([q,p])
    return pairs            
```

```python
def aliquotSequence(n):
    seq =[n]
    for i in range(200):
        p=aliquotSum(n)
        if p == 1:
            seq.append(1)
            break
        elif p == n:
            break
        else:
            seq.append(p)  
        n=p
    return seq
```

```python
#@title Select an option

selection = 'Aliquot Sequence'  #@param ["Amicable pairs","Aliquot Sum","Aliquot Sequence"]
input_number = 26 #@param {type:"integer"}
```

```python
if selection == 'Amicable pairs':
    print(f'Amicable pairs less than {input_number} are: {amicable(input_number)}')
elif selection == 'Aliquot Sum':
    print(f'The aliquot sum of {input_number} is {aliquotSum(input_number)}')
elif selection == 'Aliquot Sequence':
    print(f'The aliquot sequence of {input_number}  is {aliquotSequence(input_number)}')
```
