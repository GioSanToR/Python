```python
input_number = 26 #@param {type:"integer"}
```

```python
from matplotlib import pyplot as plt

j = int(input_number)
i = 0
y = 1
steps = []
integers = []
for x in range (2, j+1):
    y += 1 
    while x > 1:
        if x%2 == 0:
            x = x/2
        else:
            x = (3*x +1)/2
        i += 1
    integers.append(y)
    steps.append(i)
    print(f"El número {y} requiere {i} pasos")
    i=0
plt.plot(integers,steps)
plt.xlabel('Integer')
plt.ylabel('Steps')
plt.show()
```
